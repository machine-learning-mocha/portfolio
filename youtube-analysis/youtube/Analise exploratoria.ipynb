```python
import pandas as pd
import numpy as np
from collections import Counter
import operator
```

```python
gbVideos = pd.read_csv('data/GBvideos.csv', error_bad_lines=False)
```

```python
gbVideos.groupby(['channel_title']).agg({'views':'sum', 'likes': 'mean', 'dislikes': 'max'})
```

```python
gbVideos.groupby(['category_id']).agg({'views':['min', 'max'], 'likes': ['min', 'max'], 'dislikes': ['min', 'max']})
```

```python
curTags = []
for i in range(0, len(gbVideos)):   
    curTags.extend(gbVideos.tags[i].lower().split('|'))
```

```python
curTags = dict(Counter(curTags))
```

```python
sorted(curTags.items(), key=lambda x: -x[1])
```
