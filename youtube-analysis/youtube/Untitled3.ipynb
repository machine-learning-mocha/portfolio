```python
from nltk.corpus import stopwords
more_stopwords = 'é um uma uns'
more_stopwrods_list = stopwords.words('portuguese')
more_stopwrods_list.extend(more_stopwords.split())
stop_words = set(more_stopwrods_list)
```

```python
texts = [[word for word in str(document).lower().split() if word not in stop_words and word.isalpha()]
         for document in documents]
```

```python
from collections import defaultdict
```

```python
frequency = defaultdict(int)
```

```python
texts
```

```python
for text in texts:
    for token in text:
        frequency[token] += 1
```

```python
frequency
```

```python
texts = [[token for token in text if frequency[token] > 2]
         for text in texts]
```

```python
texts
```

```python
from gensim.corpora import Dictionary
dct = Dictionary(texts)
# dct.save_as_text('test.txt')
```

```python
corpus = [dct.doc2bow(line) for line in texts]
```
